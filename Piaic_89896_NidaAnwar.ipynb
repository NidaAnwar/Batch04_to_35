```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
```

```python
files=glob("Assignment/Assignment/states*")
files
```

```python
us_census = pd.concat((pd.read_csv(file) for file in files ),ignore_index=True)
us_census.head(3)
```

```python
us_census['Income']=us_census['Income'].replace('[/$]',"",regex=True)
us_census['Income'] = pd.to_numeric(us_census['Income'])
print(us_census.dtypes)
```

```python
split_gender = us_census['GenderPop'].str.split('_', expand=True)
us_census['Female'] = split_gender[1].str.split('(\d+)', expand=True)[1]
us_census['Male'] = split_gender[0].str.split('(\d+)', expand=True)[1]
us_census.head(3)
```

```python
us_census.dtypes
```

```python
us_census['Male']=us_census['Male'].replace('[/$]',"",regex=True)
us_census['Male'] = pd.to_numeric(us_census['Male'])
us_census['Female']=us_census['Female'].replace('[/$]',"",regex=True)
us_census['Female'] = pd.to_numeric(us_census['Female'])
```

```python
us_census.head(3)
```

```python
us_census.dtypes
```

```python
plt.scatter(x="Female",y="Income")
plt.show()
```

```python
us_census.head(4)
```

```python
us_census.Female.isnull().any()
```

```python
values=us_census["TotalPop"]-us_census["Male"]
us_census['Female']=us_census['Female'].fillna(value= values)
```

```python
us_census.Female.isnull().any()
```

```python
us_census.dtypes
```

```python
print(us_census["Female"].drop_duplicates())
```

```python
plt.scatter(x="Female",y="Male")
plt.scatter(x="Female",y="Income")
plt.show()
```

```python
us_census['Hispanic']=us_census['Hispanic'].replace('[/%]',"",regex=True)
us_census['Hispanic'] = pd.to_numeric(us_census['Hispanic'])
us_census['White']=us_census['White'].replace('[/%]',"",regex=True)
us_census['White'] = pd.to_numeric(us_census['White'])
us_census['Black']=us_census['Black'].replace('[/%]',"",regex=True)
us_census['Black'] = pd.to_numeric(us_census['Black'])
us_census['Native']=us_census['Native'].replace('[/%]',"",regex=True)
us_census['Native'] = pd.to_numeric(us_census['Native'])
us_census['Asian']=us_census['Asian'].replace('[/%]',"",regex=True)
us_census['Asian'] = pd.to_numeric(us_census['Asian'])
us_census['Pacific']=us_census['Pacific'].replace('[/%]',"",regex=True)
us_census['Pacific'] = pd.to_numeric(us_census['Pacific'])
```

```python
us_census=us_census.drop("GenderPop",axis=1)
us_census.head(3)
```

```python
us_census.plot(x="Female",y="Income",kind="hist",color="c",title="Female pop Verses\n Income")
```

```python
print(us_census["Male"].isnull().any())
print(us_census["Male"].drop_duplicates())
```

```python
us_census.plot(x="Male",y="Income",kind="hist",color="y",title="Male pop Verses\n Income")
```

```python
us_census["State"].isnull().any()
print(us_census["State"].drop_duplicates())
```

```python
us_census.plot(x="State",y=["TotalPop","Male","Female"],kind="hist",color=["k","m","c"],title="Total_pop Verses\n State")
```

```python
print(us_census["White"].isnull().any())
print(us_census["White"].drop_duplicates())
print(us_census["Black"].isnull().any())
print(us_census["Black"].drop_duplicates())
print(us_census["Native"].isnull().any())
print(us_census["Native"].drop_duplicates())
print(us_census["Asian"].isnull().any())
print(us_census["Asian"].drop_duplicates())
```

```python
us_census.plot(x="State",y=["White","Black","Native","Asian"],kind="hist",color=["c","g","b","y"],\
               title="State vs/ W,B,N,A")
```

```python
us_census["Hispanic"].isnull().any()
us_census["Hispanic"].drop_duplicates()
us_census["Pacific"].isnull().any()
us_census["Pacific"].drop_duplicates()
```

```python
us_census.plot(x="State",y=["Hispanic","Pacific"],kind="hist",\
               title="State vs/ Hispanic & Pacific ")
```
